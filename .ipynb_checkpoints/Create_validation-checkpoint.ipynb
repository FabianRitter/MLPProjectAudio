```python
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
categories = df_train['label'].unique()
counts = df_train['label'].value_counts().to_dict()
```

```python
df_train.head()
```

```python
plt.figure(figsize=(15,8))
ax = sns.countplot(x="label", hue="manually_verified",
                   data=df_train,order = df_train['label'].value_counts().index,
                  palette="deep")
ax.set_xticklabels(categories,rotation=90)
```

```python
plt.figure(figsize=(15,8))
ax = sns.countplot(x="label", hue="noisy_small",
                   data=df_train,order = df_train['label'].value_counts().index,
                  palette="deep")
ax.set_xticklabels(categories,rotation=90)
```

```python
counts_norm = {k: v / total for total in (sum(counts.values()),) for k, v in counts.items()}
fig, ax = plt.subplots(figsize=(12,6))
ax.bar(counts_norm.keys(), counts_norm.values())
```

```python
plt.figure(figsize=(15,8))
ax = sns.countplot(x="label",
                   data=df_test,order = df_test['label'].value_counts().index,
                  palette="deep")
ax.set_xticklabels(categories,rotation=90)
```

```python
counts_test = df_test['label'].value_counts().to_dict()
```

```python
counts_test = {k: v / total for total in (sum(counts_test.values()),) for k, v in counts_test.items()}
fig, ax = plt.subplots(figsize=(12,6))
ax.bar(counts_test.keys(), counts_test.values())
```
